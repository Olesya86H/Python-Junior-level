Задание 1. Дано натуральное число. Определить: а) количество цифр 3 в нем; б) сколько раз в нем встречается последняя цифра; в) количество четных цифр в нем

```python
a = input()
```

```python
k = 0
for i in a:
    if i == '3':
        k += 1
print('Количество цифр 3: ', k)    
```

```python
b = a[::-1]
k = 0
last_item = b[0]
for i in b:
    if i == last_item:
        k += 1
print('Последняя цифра: {f1} встречается {f2} раз(а)'.format(f1=last_item,f2=k))           
```

```python
k = 0
for i in a:    
    if (int(i) % 2) == 0:
        k += 1
print('Количество чётных цифр: ', k)    
```

Задание 2. Напишите программу, считывает натуральное число и запускает обратный отсчет.  Для каждой секунды выводит сообщение: «Осталось: n секунд.» В конце выводит сообщение: «Мотор. Камера. Поехали.»

```python
a = input()
```

```python
import time
b = int(a)
while b >= 0:
    time.sleep(1)
    print('Осталось', b)    
    b -= 1
print('Мотор. Камера. Поехали.')
```

Задание 4. Напишите программу, которая для заданного начального числа N выводит последовательность чисел по следующему правилу: если число четное, разделите его на 2; если число нечетное, умножьте его на 3 и прибавьте 1. Программа должна продолжать выполнение до тех пор, пока не будет достигнуто число 1.

```python
a = input()
```

```python
i = int(a)
if i % 2 == 0:
    while i != 1:
        i = i // 2
        print(i)
else:
    while i != 1:
        i = i*3+1
        print(i)
```

Задание 5. Напишите программу, которая считывает числа, пока не будет введено число 0. Если вводится число, которое в котором есть 7, выводится сообщение «Счастливое», если вводится число, которое содержит 22, то выводится сообщение «Опасность» и программа завершает свою работу.

```python
s = input()
is_2 = 0
is_dangerous = 0
while s != '0':
    for i in s:
        if i == '7':
            print('Счастливое')
            break
        elif i == '2':
            if is_2 == 1:
                print('Опасность')
                is_dangerous = 1
                break 
            is_2 = 1    
    if is_dangerous == 0:
        s = input()
    elif is_dangerous == 1:
        s = '0'
    is_2 = 0
    is_dangerous = 0
        
```
